```python
#** NO. OF MATCHES WON BY TEAMS**#

team_wins=ipl["winner"].value_counts()
print(team_wins)



plt.figure(figsize=(10,5))
sns.countplot(data=ipl,y="winner")
plt.grid()
plt.title("MATCHES WON COUNT")

plt.xlabel("max wins")
plt.ylabel("winner teams")
```

```python
#**NO. OF TOSS WON BY TEAMS

plt.figure(figsize=(10,5))
sns.countplot(data=ipl,y="toss_winner")
plt.xlim(0,90)
plt.xlabel("count of tosses")
plt.grid()
plt.title("TOSS WON COUNT")
plt.ylabel("TEAMS")
```

```python
#**MAN OF THE MATCH**#


man_of_the_match=ipl["player_of_match"].value_counts()
man_of_the_match.head(5)



```

```python
#** MATCHES WON BY TEAMS IN DIFFERENT LOCATIONS**#

locations=ipl[(ipl["team1"]==ipl["winner"])|(ipl["team2"]==ipl["winner"])]["city"].groupby(ipl["winner"]).value_counts()
locations
```

```python
#**TOSS WON AND NO. OF MATCHes WON OR LOST BY TEAMS UPON WINNING TOSS**#

toss_winner_match_winner=ipl[(ipl["team1"]==ipl["toss_winner"])|(ipl["team2"]==ipl["toss_winner"])]["winner"].groupby(ipl["toss_winner"]).value_counts()
toss_winner_match_winner
```

```python
#**HOME ADVANTAGE PIE CHART**#
#**ASSUME MATCH PLAYED ON GROUND IS HOME GROUND OF TEAM1**#

home_advantage=ipl[(ipl["team1"]==ipl["winner"])]["team1"].value_counts()
print("HOME ADVANTAGE TEAMS\n",home_advantage)
labels=home_advantage.index.tolist()

x=np.array(home_advantage)
plt.figure(figsize=(10,10))
plt.pie(x,labels=labels,autopct="%1.1f%%")
plt.title("TEAMS WITH HOME ADVANTAGE")
```
